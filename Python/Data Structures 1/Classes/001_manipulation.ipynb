- Manipulacao de variavel e operadores. 
Uma variavel é um espaço de memoria que guarda um valor.

- Constante
É um valor que não deveria mudar durante o programa. Usando Letrar maiusculas. Exemplos: PI = 3.14 / GRAVIDADE = 9.8

- Função type(v)
Retorna o tipo da varivael.

- Varivael numerica e string
* Int são numeros inteiros.
* float são numeros flutuntes.
* bool são true or false.
* complex é usado para numeros complexos. Usando a = parte real, b = parte imaginaria e i = unidade imaginária (√-1).

- Varivael Lista (lista = [])
* Servem para agrudar variaveis e assim transportar esses objetos de forma agrupadas. Na vida real ultilizamos caixas para transportar.

*A lista permite adicionar elementos (variavel.append())

- Tuplas (tupla = ())
* São um tipo de coleção em Python usadas para armazenar vários valores em uma única variável, parecido com listas.
*Tuplas não permite adicionar elementos depois de criada.
*A diferença entre Tuplas e Listas são que as tuplas não permitem adicionar elementos depois de criadas enquanto a lista permite voce adicionar elementos.

- Dicionarios (dict) são uma estrutura de dados que armazena informações em pares de chave e valor.
* A vantagem do dicionario é.

- Conversão de dados
* É o processo de transformar um tipo de dado em outro tipo.

*int(v) - Converte um valor para número decimal.

*float(v) - Converte um valor para número decimal.

*str(v) - Converte um valor para texto (string).

*hex(v) - Converte um número inteiro para hexadecimal (base 16).

- Operadores Aritmeticos

*Adição (+)

*Subtração (-)

*Multiplicação (*)

*Divisão (/)

*Resto da divisão (%)

*Exponenciação (**)

*Divisão Inteira (//)

- Operadores de compração

*Igualdade (==)

*Diferença (!=)

*Maior (>)

*Menor (>)

*Maior igual(>=)

*Menor igual (<=)

- Operadores de Atribuição

* = Atribuir um valor

* += Atribuir um valor somado.

* -= Atribuir um valor diminuindo.

* *= Atribuir um valor multiplicando

* /= Atribuir um valor dividindo.

* %= Atribuir um valor calculando o resto da divisão

* **= Atribuir um valor elevado a uma potencia.

* //= Atribuir um valor calculando sua divisão inteira.





```python
# Types e Variaveis numericas e strings
a = 3
print(type(a))
print(a)

print("--------")

b = 3.0
print(type(b))
print(b)

print("--------")

c = False
print(type(c))
print(c)

d = 5 + 3
print(type(d))
print(d)    

print("--------")

e = "Exemplo de string"

print(type(e))
print(e)

#Types de listas

lista = [1, 2, 3, 4, 5, "teste"] #Podemos usar variaveis numericas e strings em uma lista.

print(type(lista))
print(lista)
lista.append("mais um")
print(lista)

print("--------")

#Tuplas

tupla = (1, 2, 3, 4, 5, "teste")
print(type(tupla))
print(tupla)    

print("--------")

#Dicionario

dicionario = {}
dicionario ['nome'] = "Lucas"
dicionario ['endereco'] = "Rua teste, 123"
print(type(dicionario))
print(dicionario)
print(dicionario['nome'])
dicionario_diferente = {"nome": "Maria", "endereco": "Rua teste, 456"}
print(dicionario_diferente['nome'])

print("--------")

v1str = "10"
v2str = "20"
totStr = v1str + v2str  
print(totStr) #Concatenacao (junção) de strings.

v1int = int(v1str) #Convertendo string para inteiro.
v2int = int(v2str)  
tot = v1int + v2int
print(tot) #Soma de inteiros.

v1FltStr = 4.5
v2FltStr = 5.5
totFltStr = v1FltStr + v2FltStr
print(totFltStr) #Soma de floats.

#float

v1FltStr = float(v1FltStr) #Convertendo string para float.
v2FltStr = float(v2FltStr)
totalflt = v1FltStr + v2FltStr
print(totalflt) #Soma de floats.



```
